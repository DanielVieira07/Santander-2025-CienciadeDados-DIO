# **Load**


Adicionando uma nova coluna no df e depois atualizando essa atributo em cada cliente com a mensagem personalizada de acordo com seus dados bancario.

```python
import re

pattern = r'\*+\s*\*\*Mensagem:\*\*\s*"(.*?)"'

mensagens_marketing = re.findall(pattern, response.text, re.DOTALL)

# Verificar se o número de mensagens extraídas corresponde ao número de linhas no DataFrame
if len(mensagens_marketing) == len(df):
    df['Mensagem_Marketing'] = mensagens_marketing
    print(df[['ID_Usuario', 'Mensagem_Marketing']].to_string())
else:
    print(f"Erro: O número de mensagens extraídas ({len(mensagens_marketing)}) não corresponde ao número de clientes no DataFrame ({len(df)}).")
```
